# Basemap tutorial

This is where I keep my notes and code snippets for learning the `Basemap` library.

The tutorial can be found at https://basemaptutorial.readthedocs.io/en/latest/index.html

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
m = Basemap()

m.drawcoastlines()

plt.show()
```

```python
m = Basemap(projection='ortho', 
              lat_0=0, lon_0=0)

#Fill the globe with a blue color 
m.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
m.fillcontinents(color='coral',lake_color='aqua')

m.drawcoastlines()

plt.show()
```

Add the `projection` argument to change projection of the map. The `cyl` projection is the default projection.

```python
m = Basemap(projection='cyl')

m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawcoastlines()

plt.show()
```

Some projection requires additional argument.

```python
m = Basemap(projection='aeqd', lon_0 = 10, lat_0 = 50)
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawcoastlines()

plt.show()
```

More information about EPSG can be found at http://spatialreference.org/ref/epsg/.

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from numpy import array
from numpy import max


m = Basemap(llcrnrlon=-0.5,llcrnrlat=39.8,urcrnrlon=4.,urcrnrlat=43.,
             resolution='i', projection='tmerc', lat_0 = 39.5, lon_0 = 1)


m.readshapefile('code_examples/sample_files/lightnings', 'lightnings')

x = []
y = []
c = []

for info, lightning in zip(m.lightnings_info, m.lightnings):
    x.append(lightning[0])
    y.append(lightning[1])
    
    if float(info['amplitude']) < 0:
        c.append(-1 * float(info['amplitude']))
    else:
        c.append(float(info['amplitude']))
    
plt.figure(0)

m.drawcoastlines()
m.readshapefile('code_examples/sample_files/comarques', 'comarques')

m.hexbin(array(x), array(y))

m.colorbar(location='bottom')



plt.figure(1)

m.drawcoastlines()
m.readshapefile('code_examples/sample_files/comarques', 'comarques')

m.hexbin(array(x), array(y), gridsize=20, mincnt=1, cmap='summer', bins='log')

m.colorbar(location='bottom', format='%.1f', label='log(# lightnings)')



plt.figure(2)

m.drawcoastlines()
m.readshapefile('code_examples/sample_files/comarques', 'comarques')

m.hexbin(array(x), array(y), gridsize=20, mincnt=1, cmap='summer', norm=colors.LogNorm())

cb = m.colorbar(location='bottom', format='%d', label='# lightnings')

cb.set_ticks([1, 5, 10, 15, 20, 25, 30])
cb.set_ticklabels([1, 5, 10, 15, 20, 25, 30])




plt.figure(3)

m.drawcoastlines()
m.readshapefile('code_examples/sample_files/comarques', 'comarques')

m.hexbin(array(x), array(y), C = array(c), reduce_C_function = max, gridsize=20, mincnt=1, cmap='YlOrBr', linewidths=0.5, edgecolors='k')

m.colorbar(location='bottom', label='Mean amplitude (kA)')

plt.show()
```
